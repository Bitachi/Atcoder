```python
import matplotlib.pyplot as plt
import math
import numpy as np

# データ生成
x = np.linspace(0, 10)

#y = x + np.random.randn(100) 
a =2
b = 4
def f(x):
    return np.floor((a * x) / b) - a*np.floor(x / b)
# プロット

plt.plot(x, f(x), label="test")

# 凡例の表示
plt.legend()

# プロット表示(設定の反映)
plt.show()
plt.ylim()
```
